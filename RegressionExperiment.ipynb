```python

import numpy as np
import sklearn.datasets as ds
import sklearn.model_selection as ms
import matplotlib.pyplot as plt

path = 'E:\housing_scale.txt'

X,y = ds.load_svmlight_file(path)
X_train, X_test, y_train, y_test = ms.train_test_split(X,y)

X_train = X_train.toarray()
X_test = X_test.toarray()

tmp = np.ones((X_train.shape[0],0))
y_train = np.insert(tmp, 0, values=y_train, axis=1)

tmp = np.ones((X_test.shape[0],0))
y_test = np.insert(tmp, 0, values=y_test, axis=1)

#y_train = y_train.toarray()
#y_test = y_test.toarray()


def loss(W, X, y):
	ans = 0
	m = X.shape[0]
	for i in range(0, m):
		ans += (X[i].dot(W)-y[i])**2
	return ans/(2*m)

def update(W, X, y):
	ret = np.zeros((X.shape[1],1))
	H = X.dot(W)-y
	for i in range(0,X.shape[1]):
		ret[i] = np.sum(H * X[:, i:i+1]) / X.shape[1]
	
	return ret


import random
def train(X, y, X_test, y_test):
	X = np.insert(X, X.shape[1], values=np.ones(X.shape[0]), axis=1)
	X_test = np.insert(X_test, X_test.shape[1], values=np.ones(X_test.shape[0]), axis=1)

	alpha = 0.0001
	W = np.zeros((X.shape[1],1))
	
	for i in range(0, W.shape[0]):
		W[i] = random.uniform(-1,1);

	px = []
	l_train = []
	l_test = []
	for i in range(1, 1001):
		W -= alpha * update(W, X, y)
		px.append(i)
		loss1 = loss(W, X, y)
		loss2 = loss(W, X_test, y_test)
		print(loss1,loss2)
		l_train.append(loss1)
		l_test.append(loss2)

	plt.plot(px, l_train, label='train set')
	plt.plot(px, l_test, label='test set')
	plt.xlabel('iteration')
	plt.ylabel('loss')
	plt.legend(loc = 'upper right')
	plt.show()


train(X_train, y_train, X_test, y)

```
